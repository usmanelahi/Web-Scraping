```python
# importing the libraries
import requests
from bs4 import BeautifulSoup
import pandas as pd


#base url is used for the main website
baseURL ="http://lhr.nu.edu.pk"

#while url is for the specific page
URL = baseURL+"/faculty/"

#I have done it separately, this can also be done in one url
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python

#creating an array with names of all departments
departments=["fsc","ee","cv","fsm","ss"]
```

```python
# Need to make a dataframe with the columns needed
df = pd.DataFrame({"ID":[],"Name":[],"Department":[],"Image URL":[],"email":[],"Designation":[],"Contact No":[],"Education":[] })
```

```python

# Using a For Loop for collection of the data

for i in range(0, len(departments)):

    department =  soup.find("div",id=departments[i])
    
#    Here, I have Extracted all the attributes from the file
    name = department.find_all("h5",class_="text-center")
    img = department.find_all("img",class_="card-img-top")
    email = department.find_all("p",class_="mb-0 text-center")
    designations = department.find_all("p", class_="small text-center font-italic")
    idLink = department.find_all("a", class_="faculty-link")
    
    contacts = [];
    educations = [];
    IDs = [];
    for j in range(0, len(idLink)):
    
    #     Now extracting details from the second page
    
        detailPage = requests.get(baseURL+idLink[j]["href"])
        detailSoup = BeautifulSoup(detailPage.content, "html.parser")
        contactNo = detailSoup.find("span", class_="small").text
        contactNo = contactNo.strip()
        
        latestEducation = detailSoup.find_all("ul")
        latestEducation = latestEducation[len(latestEducation)-1]
        latestEducation = latestEducation.find("li").text.strip()

    #         Now extract id from the link
    
        ID = idLink[j]["href"]
        ID = ID.strip()
        ID = ID.split("/")
        ID = ID[3]
        
        
#       Using attribute of HEC Approved PhD Supervisor by using tokenization
    
        desigTemp = designations[0].text
        desigTemp = desigTemp.split("\n")
        desigTemp = desigTemp[0]
        
#        insert data into data frame
        df = df.append({"ID":ID.strip(),"Name":name[j].text.strip(),"Department":departments[i],"Image URL":baseURL+img[j]["src"].strip(),"email":email[j].text.strip(),"Designation":desigTemp.strip(),"Contact No":contactNo,"Education":latestEducation.strip()}, ignore_index=True)

        
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

while(1): # try to read data untill we don't get any response from the server if get break the loop and continue
    try:
        baseURL ="http://lhr.nu.edu.pk"
        URL = baseURL+"/faculty/"
        page = requests.get(URL)

        soup = BeautifulSoup(page.content, "html.parser")
        break
    except:
        print("Server is offline")
```

```python
departments=["fsc","ee","cv","fsm","ss"]
```

```python
for i in range(0, len(departments)):
    # Now we need to create a data frame
    df = pd.DataFrame({"ID":[],"Name":[],"Designation":[],"HEC Approved PhD Supervisor":[],"Email":[],"Department":[],"ImageURL":[]})
    
    department =  soup.find("div",id=departments[i])
    
#     Extract Name
    name = department.find_all("h5",class_="text-center")
#     Extract Image link
    img = department.find_all("img",class_="card-img-top")
#     Extract Email
    email = department.find_all("p",class_="mb-0 text-center")
#     Extract Designation
    designations = department.find_all("p", class_="small text-center font-italic")
#     Extract Id link
    idLink = department.find_all("a", class_="faculty-link")
    
#     Extract department name
    depName = department.find("h1", class_="mb-2 mt-3").text.strip()
    for j in range(0, len(idLink)):
        # Now extract id from the link
        ID = idLink[j]["href"]
        ID = ID.strip()
        ID = ID.split("/")
        ID = ID[3]
        
        # tokenize the extra designation like "HEC Approved PhD Supervisor" etc
        desigTemp = designations[j].text.strip()
        desigTemp = desigTemp.split("\n")
        
        # Now we need to check is there any prof HEC Approved PhD Supervisor        
        isSupervisor = False
        if(desigTemp[-1].strip() == "HEC Approved PhD Supervisor"):
            isSupervisor = True
        df = df.append({"ID":int(ID),"Name":name[j].text.strip(),"Designation":desigTemp[0].strip(),"HEC Approved PhD Supervisor":bool(isSupervisor),"Email":email[j].text.strip(),"Department":depName,"ImageURL":baseURL+img[j]["src"].strip()}, ignore_index=True)
        df.to_csv(departments[i]+".csv")
```

```python
import pandas as pd

#loading the Computing School files in dataframes
fsc = pd.read_csv ('fsc.csv')
fsc2 = pd.read_csv ('fsc_2.csv')

#putput just for checking
fsc2
```

```python
#fsc_C is the complete file for FAST School Of Computing
#applied left join based on the ID attribute (column)

fsc_c = fsc.merge(fsc2, on = 'ID', how = 'left')

#checking the output file
fsc_c

```

```python
#loading the management School files in dataframes
fsm = pd.read_csv ('fsm.csv')
fsm2 = pd.read_csv ('fsm_2.csv')

fsm_c = fsm.merge(fsm, on = 'ID', how = 'left')

#checking the output file
fsm_c


```

```python

ee = pd.read_csv ('ee.csv')
ee2 = pd.read_csv ('ee_2.csv')


ee2

ee_c = ee.merge(ee, on = 'ID', how = 'left')

#checking the output file
ee_c
```

```python
#loading the Civil Engineering School files in dataframes
cv = pd.read_csv ('cv.csv')
cv2 = pd.read_csv ('cv_2.csv')

cv2
cv_c = cv.merge(cv2, on = 'ID', how = 'left')

#checking the output file
cv_c
```

```python
#loading the Social Sciences School files in dataframes
ss = pd.read_csv ('ss.csv')
ss2 = pd.read_csv ('ss_2.csv')

ss2

ss_c = ss.merge(ss2, on = 'ID', how = 'left')

#checking the output file
ss_c
```

```python
fast_lhr_faculty = pd.concat ([fsc_c,fsm_c], axis=0)
fast_lhr_faculty = pd.concat ([fast_lhr_faculty,ee_c], axis=0)
fast_lhr_faculty = pd.concat ([fast_lhr_faculty,cv_c], axis=0)
fast_lhr_faculty = pd.concat ([fast_lhr_faculty,ss_c], axis=0)


fast_lhr_faculty
```
